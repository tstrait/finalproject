```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

Get RPM data for past 3 seasons

```python
#Get all the plus/minus data for 2017-2018, iterate through each page of stats
plusminus_topplayers = pd.read_html('http://www.espn.com/nba/statistics/rpm/_/sort/WINS')
rpm_1718 = plusminus_topplayers[0]

for i in range(2,15):
    rpm_url = 'http://www.espn.com/nba/statistics/rpm/_/page/' + str(i) + '/sort/WINS'
    plusminuspage = pd.read_html(rpm_url)
    rpm_1718 = pd.concat([rpm_1718, plusminuspage[0]])

#Get all the plus/minus data for 2016-2017, iterate through each page of stats
plusminus_topplayers = pd.read_html('http://www.espn.com/nba/statistics/rpm/_/year/2017/sort/WINS')
rpm_1617 = plusminus_topplayers[0]

for i in range(2,13):
    rpm_url = 'http://www.espn.com/nba/statistics/rpm/_/year/2017/page/' + str(i) + '/sort/WINS'
    plusminuspage = pd.read_html(rpm_url)
    rpm_1617 = pd.concat([rpm_1617, plusminuspage[0]])
    
#Get all the plus/minus data for 2015-2016, iterate through each page of stats
plusminus_topplayers = pd.read_html('http://www.espn.com/nba/statistics/rpm/_/year/2016/sort/WINS')
rpm_1516 = plusminus_topplayers[0]

for i in range(2,12):
    rpm_url = 'http://www.espn.com/nba/statistics/rpm/_/year/2016/page/' + str(i) + '/sort/WINS'
    plusminuspage = pd.read_html(rpm_url)
    rpm_1516 = pd.concat([rpm_1516, plusminuspage[0]])
    
rpm_1617.reset_index()
rpm_1718.reset_index()
rpm_1516.reset_index()
```

```python
#post to CSV, clean things up in excel to have a clean data source for other notebooks
rpm_1516.to_csv('rpm1516.csv')
rpm_1617.to_csv('rpm1617.csv')
rpm_1718.to_csv('rpm1718.csv')
```

Salary Data

```python
salaries = pd.read_html('http://www.espn.com/nba/salaries/_/seasontype/3')
salaries_1718 = salaries[0]

for i in range(2,14):
    sal_url = 'http://www.espn.com/nba/salaries/_/page/' + str(i) + '/seasontype/3'
    salariesdata = pd.read_html(sal_url)
    salaries_1718 = pd.concat([salaries_1718, salariesdata[0]])

#Get all the salary data for 2016-2017, iterate through each page of stats
salaries = pd.read_html('http://www.espn.com/nba/salaries/_/year/2017/seasontype/3')
salaries_1617 = salaries[0]

for i in range(2,13):
    sal_url = 'http://www.espn.com/nba/salaries/_/year/2017/page/' + str(i) + '/seasontype/3'
    salariesdata = pd.read_html(sal_url)
    salaries_1617 = pd.concat([salaries_1617, salariesdata[0]])
    
#Get all the salary data for 2015-2016, iterate through each page of stats
salaries = pd.read_html('http://www.espn.com/nba/salaries/_/year/2016/seasontype/3')
salaries_1516 = salaries[0]

for i in range(2,12):
    sal_url = 'http://www.espn.com/nba/salaries/_/year/2016/page/' + str(i) + '/seasontype/3'
    salariesdata = pd.read_html(sal_url)
    salaries_1516 = pd.concat([salaries_1516, salariesdata[0]])
```

```python
#post to CSV, clean things up in excel to have a clean data source for other notebooks
salaries_1516.to_csv('salaries_1516.csv')
salaries_1617.to_csv('salaries_1617.csv')
salaries_1718.to_csv('salaries_1718.csv')
```
